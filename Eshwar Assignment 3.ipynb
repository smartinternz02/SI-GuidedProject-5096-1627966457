```python
import seaborn as sns
import numpy as np
```

```python
data_iris = sns.load_dataset("iris")
```

```python
data_iris.head()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = data_iris)
```

The scatter plot shows that data points with large sepal width but lower sepal length seem to be clustered together and vice versa

```python
sns.lineplot(x = "sepal_length",y="petal_length",data = data_iris,ci=None)
```

The lineplot shows that that petal_length and sepal_legth of iris have positive corelation

```python
sns.distplot(data_iris['petal_length'])
```

Displot has been used to show the variation in data distribution of the petal_length. it can be seen that the data has highest density from 4-6.

```python
sns.barplot(x= "species",y = "sepal_width",data = data_iris)
```

```python
sns.countplot(x= "species",data = data_iris)
```

The boxplot shows the total number of each species

```python
sns.boxplot(x = "species",y = "petal_length",data = data_iris)
```

The boxplot shows that the species of Iris can be clearly categorized using the petal length

```python
sns.catplot(x="species",y="sepal_length",data=data_iris)
```

The catplot shows that the different species of iris have differernt sepal_length i.e., the setosa species has the lowest sepal length and virginica has the highest sepal_length.

```python
data_iris.corr()
```

```python
sns.heatmap(data_iris.corr(),annot = True, cmap="YlGnBu")
```

The heatmap has been used to show the corelation between sepal_length, sepal_width, petal_length and petal_width
